```python
# create a function to clear outliers from the data.

def cleanColumn(data, columns, thr = 1):
  column_desc = describe[columns]

  q1 = column_desc[4]
  q3 = column_desc[6]
  IQR = q3 - q1 # Inter Quartile Range

  top_limit = q3 + thr * IQR
  bottom_limit = q1 - thr * IQR

  bottom_limit_rows = bottom_limit < data[columns]
  top_limit_rows = data[columns] < top_limit

  filters = bottom_limit_rows & top_limit_rows

  data = data[filters]

  return data

```

```python
data_copy = data.copy()

for i in data_copy.columns:
  data_copy = cleanColumn(data_copy, i, 2)
  print(f'{data_copy.shape} after column {i} ')
```
